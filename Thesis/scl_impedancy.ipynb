The goal of this Notebook is to show how impedance from a space charge model differs from the experimentally measured grain boundary impedance.

```python
import numpy as np
from matplotlib import pyplot as plt
from cmath import sqrt
```

## Initialization

The experimental data is impedance spectroscopy data from a 10 mol% doped BZY sample, acquired at T=523 K.

```python
#Importing experimental data
freq_data = np.genfromtxt('freq.csv', delimiter=',')
z_real_data = np.genfromtxt('Re Z.csv', delimiter=',')
z_imag_data = np.genfromtxt('Im Z.csv', delimiter=',')
w_data = freq_data*2*np.pi
```

The impedance spectroscopy data was fitted the conventional way with a constant phase element, yielding the following parameters.

```python
R_gb_expt = 18612 #Grain boundary resistance
Y_gb_expt = 0.000000268 #Constant phase element param
n_gb_expt = 0.66509 #Constant phase element param
C_gb_expt = 1.85743126797478e-8 #Grain boundary capacitance
rho_bulk_expt = 54.6637121724624 #Bulk resistivity
w0_bulk_expt = 3.5e7 #Bulk characteristic frequency
C_bulk_expt = 4.1e-11 #Bulk capacitance
epsilon = 1/(rho_bulk_expt*w0_bulk_expt) #Dielectric constant
R_gb_bulk = 696
```

```python
#Sample geometry:
sample_area = (10e-3)**2*np.pi
sample_length = 1e-3
grain_size = 0.5e-6
n_GBs = sample_length/grain_size #Number of grain boundaries
n_SCLs = 2*n_GBs #Number of space charge layers, accounting for both sides of the GB
```

```python
#Material specific parameters:
a_0 = 4.2e-10 #Lattice parameter
c_acc = 0.1 #Acceptor doping concentration, mol%
c_acc_m3 = c_acc/a_0**3
c_oh_bulk = c_acc
```

```python
#Modelling parameters:
j = sqrt(-1)
w_list = np.logspace(-6, 12, 4000)
T = 523
phi_0_iterator = 1e-4
x_steps_scl = 1000
```

```python
#Physical constants
e = 1.6e-19
kbeV = 8.617e-5
```

Functions needed to model the impedance

```python
def get_expt_impedance():
    return R_gb_expt/(1 + (j*w_list)**n_gb_expt*R_gb_expt*Y_gb_expt)

def get_scl_length(phi_0):
    return sqrt((2*epsilon*phi_0)/(e*c_acc_m3))

def get_phi_scl(phi_0):
    scl_length = get_scl_length(phi_0)
    x_range = np.linspace(0, scl_length, x_steps_scl)
    phi_scl = phi_0*(x_range/scl_length - 1)**2
    return x_range.real, phi_scl.real

def get_rho_scl(phi_scl):
    return rho_bulk_expt*np.exp((phi_scl)/(kbeV*T))

def get_impedance_scl(x_range, rho_scl):
    delta_x = x_range[1] - x_range[0]
    z_micro_model = []
    for w in w_list:
        z_micro_model.append(np.sum((rho_scl*delta_x)/(1 + j*w*rho_scl*epsilon)))
    z_micro_model = (n_SCLs/sample_area)*np.asarray(z_micro_model)
    return z_micro_model
```

## Modelling

Now we are ready to model the impedance. First, we will model the impedance with a constant phase element, based on the fitting parameters obtained when fitting the experimental data to three RQ elements connected in parallel (where Q stands for constant phase element).

```python
z_macro_model = get_expt_impedance()
```

Now we model the space charge layer impedance. Since we in the space charge model assumes that the entire grain boundary resistance stems from space charge layers, we set as constraint in the modelling that the total space charge layer resistance must be equal to the experimentally measured grain boundary resistance.

This is done by setting a too low initial grain boundary core potential $V_E(0)$ ("phi_0" in the code), and increasing it incrementally until the modelled space charge layer resistance equals the experimental grain boundary resistance.

```python
phi_0 = 0.1 #GB core potential at first iteration
R_gb_modelled = -1 #Dummy value
while R_gb_modelled.real < z_macro_model.real[0]:
    scl_length = get_scl_length(phi_0)
    x_range, phi_scl = get_phi_scl(phi_0)
    rho_scl = get_rho_scl(phi_scl)
    R_gb_modelled = (n_SCLs/sample_area)*scl_length*np.average(rho_scl)
    phi_0 = phi_0 + phi_0_iterator
z_micro_model = get_impedance_scl(x_range, rho_scl)
```

## Visualization

First, we will show that the resistivity varies with 5 orders of magnitude in the space charge layer.

```python
ax = plt.gca()
ax.plot(x_range, rho_scl/rho_bulk_expt)
ax.set_xlabel("x / m", size=14)
ax.set_ylabel(r"$\frac{\rho_{gb}}{\rho_{bulk}}$", size=22)
ax.set_yscale("log")
plt.show()
```

Now, let us compare the Nyquist plots between the experimental data, the modelled data based on the macroscopic fitting parameters, and the modelled data based on the space charge model. For low frequencies, the experimental data deviates from the model, because contributions from the electrode semicircle start becoming prominent.

The constant phase element model gets an excellent fit to the experimental data. At first glance, the fit of the space charge model looks reasonably well. However, the Nyquist plot hide large deviations in the frequency dependency, as we will see in our next plot.

```python
plt.plot(z_real_data-R_gb_bulk, -z_imag_data, linestyle='dotted', label="Experimental data")
plt.plot(z_macro_model.real, -z_macro_model.imag, label="Constant phase element, model")
plt.plot(z_micro_model.real, -z_micro_model.imag, label="Space charge model")
plt.xlim(0, 2e4)
plt.ylim(0, 1e4)
plt.xlabel("Z'")
plt.ylabel("-Z''")
plt.legend(loc="best")
plt.show()
```

The plot below shows $-Z''$ vs $\omega$ for the experimental data and the two models. In this plot, it becomes evident that the space charge model has a frequency dependency one order of magnitude off compared to the experimental data. The constant phase element, on the other hand, has an almost perfect fit. Again, the experimental data deviates strongly at low frequencies, due to contributions from the electrode. 

```python
plt.plot(w_data, -z_imag_data, linestyle='dotted', label="Experimental data")
plt.plot(w_list, -z_macro_model.imag, label="Constant phase element, model")
plt.plot(w_list, -z_micro_model.imag, label="Space charge model")
plt.xscale("log")
plt.xlim(1, 1e7)
plt.ylim(0, 1.15e4)
plt.legend(loc="best")
plt.show()
```
